# Surface load DRM

This is a project creating surface load using analytical soultion for verifying the DRM and PML. 

The core code is DRM.m file which create disp.txt, acc.txt, coords.txt and vel.txt written by Adriano Throno. 

here is the schematic of the problem we want to create the load surwace load in a DRM format on the bounadries.

![shematic of problem](./../Single/figs/schematic.png)






Then the ouput files will be used by the H5DRMcreator.py file for creatinfg DRM loadind dataset in HDf5 format in the format that OpenSees can use. 



The DRM.m file also creates the controlpoints displacements and accelartions files for verifying numerical simulations. 

Here is the acclerations and dispalcements for the control points 

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3,1,sharex=True)

controldispx = np.loadtxt("disxpcontrol.txt",delimiter=",")
controldispz = np.loadtxt("disypcontrol.txt",delimiter=",")
ax[0].plot(controldispx[0,:],controldispx[[1,2,3],:].T,"--")
ax[1].plot(controldispz[0,:],controldispz[[1,2,3],:].T*0,"--")
ax[2].plot(controldispz[0,:],controldispz[[1,2,3],:].T,"--")
ax[2].set_xlabel("Time [s]")
ax[0].set_ylabel("X disp [m]")
ax[1].set_ylabel("Y disp [m]")
ax[2].set_ylabel("Z disp [m]")
ax[0].set_ylim([-0.011, 0.011])
ax[1].set_ylim([-0.011, 0.011])
ax[0].grid(True, which="both", ls="--")
ax[1].grid(True, which="both", ls="--")
ax[2].grid(True, which="both", ls="--")
ax[0].legend(["P1","P2","P3"])
```
